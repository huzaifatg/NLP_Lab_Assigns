```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('omw-1.4')
```

```python
text = "the cat is loafing on the mat"
```

**Whitespace Tokenization**

```python
from nltk.tokenize import WhitespaceTokenizer

if 'text' not in globals():
    text = "the cat is loafing on the mat"

wt = WhitespaceTokenizer()
tokens_whitespace = wt.tokenize(text)
print("Whitespace Tokenization:")
print(tokens_whitespace)
```

**Punctuation-Based Tokenization**

```python
from nltk.tokenize import wordpunct_tokenize

tokens_punct = wordpunct_tokenize(text)
print("\nPunctuation-based Tokenization:")
print(tokens_punct)
```

**Treebank Tokenization**

```python
from nltk.tokenize import TreebankWordTokenizer

twt = TreebankWordTokenizer()
tokens_treebank = twt.tokenize(text)
print("\nTreebank Tokenization:")
print(tokens_treebank)
```

**Tweet Tokenization**

```python
from nltk.tokenize import TweetTokenizer

tt = TweetTokenizer()
tokens_tweet = tt.tokenize(text)
print("\nTweet Tokenization:")
print(tokens_tweet)
```

**Multi-Word Expression (MWE) Tokenization**

```python
from nltk.tokenize import MWETokenizer

mwe = MWETokenizer([('New', 'York'), ('machine', 'learning')], separator='_')
tokens_mwe = mwe.tokenize(text.split())
print("\nMWE Tokenization:")
print(tokens_mwe)
```

**Porter Stemmer**

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()
porter_stems = [ps.stem(word) for word in tokens_treebank]

print("\nPorter Stemmer:")
print(porter_stems)
```

**Snowball Stemmer**

```python
from nltk.stem import SnowballStemmer

ss = SnowballStemmer("english")
snowball_stems = [ss.stem(word) for word in tokens_treebank]

print("\nSnowball Stemmer:")
print(snowball_stems)
```

**Lemmatization using WordNet Lemmatizer**

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in tokens_treebank]

print("\nLemmatization:")
print(lemmatized_words)
```
